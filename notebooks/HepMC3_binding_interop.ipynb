# Using HepMC3 python tools with NUISANCE event sources

```python
import pyNUISANCE as pn
```

```python
evs = pn.EventSource("nuwro.event1.dune_argon_sf_10mega.root")
if not evs:
    print("Error: failed to open input file")

# we can get HepMC3 to print out the run information for the file we just opened
run_info = evs.first()[0].run_info()
pn.hm.HepMC3.Print.listing(run_info) # this is quite loud
```

## Event information

```python
i = 0
for ev, _ in evs:
    # we can use the HepMC3 python bindings to help prettyprint events
    pn.hm.HepMC3.Print.listing(ev)
    if i == 4: # print info from the first 5 events
        break
    i += 1
```
